```python
import pandas as pd
import numpy as np
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\hency\OneDrive\Desktop\Internship\day 7\multiLinear\insurance.csv')
df1 = pd.DataFrame(data)
print(df1)
```

```python
df1.isnull()
```

```python
df3 = df1.dropna()
print(df3)
```

```python
print(df3.duplicated())
df3 = df3.drop_duplicates()
print(df3)
```

```python
plt.scatter(df3['age'],df3['charges'],color='red')
plt.show()
df4 = df3
```

```python
regr = linear_model.LinearRegression()
x = df4[['age','bmi']]
y = np.asanyarray(df4['charges'])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=0)
print("X_train:",X_train.shape)
print("X_test:",X_test.shape)
print("Y_train:",y_train.shape)
print("Y_test:",y_test.shape)
Y = regr.fit(X_train,y_train)
print(X_test)
y_pred = regr.predict(X_test)
print(y_pred)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
b = regr.predict([[19,27.9]])
print("Charges: {}".format(b))
plt.scatter(y_test,y_pred,color='g')
plt.xlabel('Actual')
plt.ylabel('Predicted')
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red')
plt.show()
```
