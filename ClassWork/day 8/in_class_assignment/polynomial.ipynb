```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Real Estate - Real estate.csv")  # Replace "data.csv" with your CSV file's path
X = data['X4 number of convenience stores'].values.reshape(-1, 1)
y = data['Y house price of unit area'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
degree = 5  # Set the degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_poly, y_train)
```

```python
y_train_pred = model.predict(X_train_poly)
y_test_pred = model.predict(X_test_poly)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X_train, y_train_pred, color='red', label='Train Prediction')
plt.plot(X_test, y_test_pred, color='green', label='Test Prediction')
plt.legend(loc='upper left')
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'Polynomial Regression (Degree {degree})')
plt.show()
```
